## Final Challenge

#### In every Olympics, which US Athlete has won the most total numbers of medals? Include Athelets discipline.


```python
gy = oo[oo.NOC == 'USA'].groupby(['Edition', 'Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
gy['Total'] = gy['Gold'] + gy['Silver'] + gy['Bronze']
gy.reset_index(inplace=True)

for year, group in gy.groupby('Edition'):
    print(group.sort_values('Total', ascending=False)[:1])
```

```python
## Want to sort them in list - using another command 

gy = oo[oo.NOC == 'USA'].groupby(['Edition', 'Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
gy['Total'] = gy['Gold'] + gy['Silver'] + gy['Bronze']
gy.reset_index(inplace=True)

tu = [ group.sort_values('Total', ascending=False)[:1] for year, group in gy.groupby('Edition')]
tu

top = pd.DataFrame()
for i in tu:
    top = top.append(i)
    
top
```
